Stochastic Gradient Descent (30 points)
In class we covered the baseline stochastic gradient descent. Using the linear regression example from the class notes, develop from scratch the baseline SGD algorithm. :

Clearly state the hyper-parameters you used and present the loss vs epoch plot that demonstrates the convergence of the algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt

#hyperparameters
learning_rate: 0.001 #Step size for weight updates.
max_iter: 1000 #Maximum number of iterations (epochs).
tol: 1e-4 #Tolerance for stopping criteria
shuffle: True #Whether to shuffle data before each epoch.
random_state: 123 #Seed for reproducibility

class LinearRegressionSGD:
    def __init__(self, learning_rate=0.001, max_iter=1000, tol=1e-4, shuffle=True, random_state=None):
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.tol = tol #Tolerance for stopping criteria
        self.shuffle = shuffle
        self.random_state = random_state
        self.weights = None #parameters to be learned
        self.bias = None #parameters to be learned
        self.loss_history = [] # store the loss values for each epoch.

    def fit(self, X, y): #This method trains the model using SGD.
        np.random.seed(self.random_state)
        self.weights = np.random.randn(X.shape[1]) #Initializes weights randomly.
        self.bias = np.random.randn() #Initializes  bias randomly.
        n_samples = X.shape[0]
        prev_loss = float('inf')
        for epoch in range(self.max_iter): #Iterates through a maximum of max_iter /epochs.
        #Optionally shuffles data at the start of each epoch
            if self.shuffle:
                X, y = self._shuffle_data(X, y)
            total_loss = 0
            #For each data point:Predicts the target value, Computes the error, Calculates and accumulates the loss, Updates the weights and bias
            for i in range(n_samples):
                sample = X[i]
                target = y[i]
                prediction = self._predict(sample)
                error = prediction - target
                total_loss += self._calculate_loss(error)
                self._update_weights(sample, error)
            avg_loss = total_loss / n_samples  #Computes the average loss for the epoch and stores it.
            self.loss_history.append(avg_loss)
            #Checks if the change in loss is less than tol to stop early.
            if np.abs(prev_loss - avg_loss) < self.tol:
                break
            prev_loss = avg_loss

    def _shuffle_data(self, X, y): #Shuffles the data by generating a random permutation of indices and applying it to both X and y.
        perm = np.random.permutation(len(X))
        return X[perm], y[perm]

    def _predict(self, X): #Computes the prediction by taking the dot product of input features X with weights and adding the bias.
        return np.dot(X, self.weights) + self.bias

    def _calculate_loss(self, error): #Computes the loss for a single prediction using the mean squared error formula(MSE)
        return 0.5 * error ** 2

    def _update_weights(self, X, error): #Updates the weights and bias based on the computed error and learning rate.
        self.weights -= self.learning_rate * error * X
        self.bias -= self.learning_rate * error

    def predict(self, X): #Public method to predict values for given input data X.
        return self._predict(X)

# Define polynomial features function
#transforms input data x into polynomial features up to the specified degree.
def polynomial_features(x, degree):
    return np.array([x ** i for i in range(1, degree + 1)]).T

# Generate toy data
def create_toy_data(func, sample_size, std, domain=[0, 1]):
    x = np.linspace(domain[0], domain[1], sample_size) #Generates sample_size evenly spaced x values within domain.
    np.random.shuffle(x) #Shuffles the x values.
    y = func(x) + np.random.normal(scale=std, size=x.shape) #Computes y values by applying func to x and adding Gaussian noise with standard deviation std.
    return x, y

def sinusoidal(x): #Defines the sinusoidal function sin(2πx).
    return np.sin(2 * np.pi * x)

#Data Generation and Transformation
x_train, y_train = create_toy_data(sinusoidal, 10, 0.25) #generate 10 train datapoints with noise
x_test = np.linspace(0, 1, 100) #Creates 100 test points without noise.
y_test = sinusoidal(x_test)

# Plot toy data
#Visualizes the training data points and the true sinusoidal function.
plt.figure(figsize=[5,3])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, label="$\sin(2\pi x)$")
plt.legend()
plt.show()

# Polynomial feature transformation
M = 9 #degree of polynomial features
#Transforms training and test data into polynomial features of degree M=9.
X_train = polynomial_features(x_train, M)
X_test = polynomial_features(x_test, M)

# Instantiate and fit the SGD model
#Model Training and Loss Plot
sgd_model = LinearRegressionSGD(learning_rate=0.001, max_iter=1000, tol=1e-4, shuffle=True, random_state=123) #Creates an instance of LinearRegressionSGD with specified hyperparameters.
sgd_model.fit(X_train, y_train) #Fits the model to the training data.

# Plot the loss vs epoch
#Plots the loss vs. epoch to demonstrate the convergence of the SGD algorithm
plt.figure(figsize=(5, 3))
plt.plot(range(len(sgd_model.loss_history)), sgd_model.loss_history)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss vs Epoch for Linear Regression with SGD')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#hyperparameters
learning_rate: 0.001 #Step size for weight updates.
max_iter: 1000 #Maximum number of iterations (epochs).
tol: 1e-4 #Tolerance for stopping criteria
shuffle: True #Whether to shuffle data before each epoch.
random_state: 123 #Seed for reproducibility

class LinearRegressionSGD:
    def __init__(self, learning_rate=0.001, max_iter=1000, tol=1e-4, shuffle=True, random_state=None):
        # Initialize hyperparameters
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.tol = tol
        self.shuffle = shuffle
        self.random_state = random_state
        self.weights = None
        self.bias = None
        self.train_loss_history = []  # To store training loss at each epoch
        self.test_loss_history = []   # To store test loss at each epoch (if provided)

    def fit(self, X_train, y_train, X_test=None, y_test=None):
        np.random.seed(self.random_state)
        # Initialize weights and bias randomly
        self.weights = np.random.randn(X_train.shape[1])
        self.bias = np.random.randn()
        n_samples = X_train.shape[0]
        prev_loss = float('inf')

        # Training loop
        for epoch in range(self.max_iter):
            if self.shuffle:
                X_train, y_train = self._shuffle_data(X_train, y_train)

            total_loss = 0
            for i in range(n_samples):
                sample = X_train[i]
                target = y_train[i]
                prediction = self._predict(sample)
                error = prediction - target
                total_loss += self._calculate_loss(error)
                self._update_weights(sample, error)

            avg_loss = total_loss / n_samples
            self.train_loss_history.append(avg_loss)

            # Calculate test loss if test data is provided
            if X_test is not None and y_test is not None:
                test_loss = self._calculate_total_loss(X_test, y_test)
                self.test_loss_history.append(test_loss)

            # Early stopping if loss change is below tolerance
            if np.abs(prev_loss - avg_loss) < self.tol:
                break
            prev_loss = avg_loss

    def _shuffle_data(self, X, y):
        # Shuffle the data
        perm = np.random.permutation(len(X))
        return X[perm], y[perm]

    def _predict(self, X):
        # Predict the output for given input X
        return np.dot(X, self.weights) + self.bias

    def _calculate_loss(self, error):
        # Calculate the loss (mean squared error)
        return 0.5 * error ** 2

    def _update_weights(self, X, error):
        # Update the weights and bias using the error and learning rate
        self.weights -= self.learning_rate * error * X
        self.bias -= self.learning_rate * error

    def _calculate_total_loss(self, X, y):
        # Calculate the total loss on the given data
        predictions = self._predict(X)
        errors = predictions - y
        total_loss = np.sum(0.5 * errors ** 2) / len(y)
        return total_loss

    def predict(self, X):
        # Predict the output for given input X
        return self._predict(X)

# Define polynomial features function
def polynomial_features(x, degree):
    # Create polynomial features up to the specified degree
    return np.array([x ** i for i in range(1, degree + 1)]).T

# Generate toy data
def create_toy_data(func, sample_size, std, domain=[0, 1]):
    x = np.linspace(domain[0], domain[1], sample_size)
    np.random.shuffle(x)
    y = func(x) + np.random.normal(scale=std, size=x.shape)
    return x, y

def sinusoidal(x):
    # Define the sinusoidal function
    return np.sin(2 * np.pi * x)

# Generate training and test data
x_train, y_train = create_toy_data(sinusoidal, 10, 0.25)
x_test = np.linspace(0, 1, 100)
y_test = sinusoidal(x_test)

# Plot toy data
plt.figure(figsize=[5, 3])
plt.scatter(x_train, y_train, facecolor="none", edgecolor="b", s=50, label="training data")
plt.plot(x_test, y_test, label="$\sin(2\pi x)$")
plt.legend()
plt.show()

# Polynomial feature transformation
M = 9
X_train = polynomial_features(x_train, M)
X_test = polynomial_features(x_test, M)

# Instantiate and fit the SGD model
sgd_model = LinearRegressionSGD(learning_rate=0.001, max_iter=1000, tol=1e-4, shuffle=True, random_state=123)
sgd_model.fit(X_train, y_train, X_test, y_test)

# Plot the loss vs epoch
plt.figure(figsize=(5, 3))
plt.plot(range(len(sgd_model.train_loss_history)), sgd_model.train_loss_history, label='Train Loss')
if sgd_model.test_loss_history:
    plt.plot(range(len(sgd_model.test_loss_history)), sgd_model.test_loss_history, label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss vs Epoch for Linear Regression with SGD')
plt.legend()
plt.grid(True)
plt.show()
```

#SGD with Momentum

Here are a 3 reasons why the loss curve for SGD with momentum may be smaller than the other two:

1. Acceleration: The momentum term allows the optimizer to accelerate along directions of persistent gradients, helping to overcome small local minima or saddle points more effectively.

2. Efficient Gradient Updates: The momentum term smooths out the variations in the gradient updates, leading to more stable and efficient updates of the model parameters.

3. Faster Convergence: Due to the acceleration and more efficient updates, SGD with momentum can converge faster to a lower loss compared to standard SGD or Adam.




```python
class LinearRegressionSGDMomentum:
    def __init__(self, learning_rate=0.001, max_iter=1000, tol=1e-4, shuffle=True, random_state=None, momentum=0.9):
        """
        Initialize the SGD with Momentum Linear Regression model.

        Parameters:
        - learning_rate (float): The learning rate for gradient descent: 0.001
        - max_iter (int): Maximum number of iterations for training: 1000
        - tol (float): Tolerance to determine convergence: 1e-4
        - shuffle (bool): Whether to shuffle the data before each epoch: True
        - random_state (int): Random seed for reproducibility.
        - momentum (float): Momentum parameter to accelerate gradient updates.0.9
        """
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.tol = tol
        self.shuffle = shuffle
        self.random_state = random_state
        self.momentum = momentum
        self.weights = None
        self.bias = None
        self.velocity_weights = None
        self.velocity_bias = None
        self.loss_history = []

    def fit(self, X, y):
        """
        The fit method trains the linear regression model using SGD with momentum.
        It iterates over the dataset for a maximum number of iterations or until convergence.
        Within each epoch, if specified, it shuffles the data. For each sample in the shuffled dataset,
        it computes the prediction using the current model parameters, calculates the prediction error,
        updates the total loss, and updates the model parameters using the momentum-based update rule

        """
        np.random.seed(self.random_state)
        self.weights = np.random.randn(X.shape[1])
        self.bias = np.random.randn()
        self.velocity_weights = np.zeros_like(self.weights)
        self.velocity_bias = 0
        n_samples = X.shape[0]
        prev_loss = float('inf')
        for epoch in range(self.max_iter):
            if self.shuffle:
                X, y = self._shuffle_data(X, y)
            total_loss = 0
            for i in range(n_samples):
                sample = X[i]
                target = y[i]
                prediction = self._predict(sample)
                error = prediction - target
                total_loss += self._calculate_loss(error)
                self._update_weights(sample, error)
            avg_loss = total_loss / n_samples
            self.loss_history.append(avg_loss)
            if np.abs(prev_loss - avg_loss) < self.tol:
                break
            prev_loss = avg_loss
        return self

    def _shuffle_data(self, X, y):
        """
        shuffles the input features and target values to introduce randomness in the training process.
        This helps prevent the model from getting biased by the order of samples during training.
        """
        perm = np.random.permutation(len(X))
        return X[perm], y[perm]

    def _predict(self, X):
        """
        computes the predicted target values using the current model parameters and the input features
        """
        return np.dot(X, self.weights) + self.bias

    def _calculate_loss(self, error):
        """
        computes the loss (mean squared error) based on the prediction error
        """
        return 0.5 * error ** 2

    def _update_weights(self, X, error):
        """
        method updates the model parameters (weights and bias) using the momentum-based update rule.
        It computes the gradients of the loss with respect to the model parameters, scales them by the learning rate,
        and updates the velocity terms. Then, it updates the model parameters using the momentum-based update rule.
        """
        grad_weights = error * X
        grad_bias = error
        self.velocity_weights = self.momentum * self.velocity_weights - self.learning_rate * grad_weights
        self.velocity_bias = self.momentum * self.velocity_bias - self.learning_rate * grad_bias
        self.weights += self.velocity_weights
        self.bias += self.velocity_bias

    def predict(self, X):
        """
        computes the predicted target values using the trained model.
        """
        return self._predict(X)
```

#Adam Optimizer

```python
class LinearRegressionAdam:
    def __init__(self, learning_rate=0.001, max_iter=1000, tol=1e-4, shuffle=True, random_state=None, beta1=0.9, beta2=0.999, epsilon=1e-8):
        """
        Initialize the Adam Linear Regression model.

        Parameters:
        - learning_rate (float): The learning rate for gradient descent: 0.001
        - max_iter (int): Maximum number of iterations for training. 1000
        - tol (float): Tolerance to determine convergence.1e-4
        - shuffle (bool): Whether to shuffle the data before each epoch.
        - random_state (int): Random seed for reproducibility.
        - beta1 (float): Exponential decay rate for the first moment estimates. 0.9
        - beta2 (float): Exponential decay rate for the second moment estimates. 0.999
        -These moments are used to calculate the adaptive learning rates for each parameter during training.
        - epsilon (float): Small value to prevent division by zero.1e-8
        """
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.tol = tol
        self.shuffle = shuffle
        self.random_state = random_state
        self.beta1 = beta1
        self.beta2 = beta2
        self.epsilon = epsilon
        self.weights = None
        self.bias = None
        self.m_t = None
        self.v_t = None
        self.m_t_bias = None
        self.v_t_bias = None
        self.loss_history = []

    def fit(self, X, y):
        """
        The fit method trains the linear regression model using the Adam optimization algorithm.
        It iterates over the dataset for a maximum number of iterations or until convergence.
        Within each epoch, if specified, it shuffles the data. For each sample in the shuffled dataset,
        it computes the prediction using the current model parameters, calculates the prediction error,
        updates the total loss, and updates the model parameters using the Adam optimization update rule.
        """
        np.random.seed(self.random_state)
        self.weights = np.random.randn(X.shape[1])
        self.bias = np.random.randn()
        self.m_t = np.zeros_like(self.weights)
        self.v_t = np.zeros_like(self.weights)
        self.m_t_bias = 0
        self.v_t_bias = 0
        n_samples = X.shape[0]
        prev_loss = float('inf')
        for epoch in range(self.max_iter):
            if self.shuffle:
                X, y = self._shuffle_data(X, y)
            total_loss = 0
            for i in range(n_samples):
                sample = X[i]
                target = y[i]
                prediction = self._predict(sample)
                error = prediction - target
                total_loss += self._calculate_loss(error)
                self._update_weights(sample, error, epoch + 1)
            avg_loss = total_loss / n_samples
            self.loss_history.append(avg_loss)
            if np.abs(prev_loss - avg_loss) < self.tol:
                break
            prev_loss = avg_loss
        return self

    def _shuffle_data(self, X, y):
        """
        shuffles the input features and target values to introduce randomness in the training process,
        preventing the model from being biased by the order of samples during training.
        """
        perm = np.random.permutation(len(X))
        return X[perm], y[perm]

    def _predict(self, X):
        """
        computes the predicted target values using the current model parameters and the input features.
        """
        return np.dot(X, self.weights) + self.bias

    def _calculate_loss(self, error):
        """
        computes the loss (mean squared error) based on the prediction error.
        """
        return 0.5 * error ** 2

    def _update_weights(self, X, error, t):
        """
        The _update_weights method updates the model parameters (weights and bias) using the Adam optimization update rule.
        It computes the gradients of the loss with respect to the model parameters,
        scales them by the learning rate and moment estimates, computes bias-corrected moment estimates,
        and updates the model parameters accordingly.
        """
        grad_weights = error * X
        grad_bias = error
        self.m_t = self.beta1 * self.m_t + (1 - self.beta1) * grad_weights
        self.v_t = self.beta2 * self.v_t + (1 - self.beta2) * (grad_weights ** 2)
        self.m_t_bias = self.beta1 * self.m_t_bias + (1 - self.beta1) * grad_bias
        self.v_t_bias = self.beta2 * self.v_t_bias + (1 - self.beta2) * (grad_bias ** 2)

        m_t_hat = self.m_t / (1 - self.beta1 ** t)
        v_t_hat = self.v_t / (1 - self.beta2 ** t)
        m_t_bias_hat = self.m_t_bias / (1 - self.beta1 ** t)
        v_t_bias_hat = self.v_t_bias / (1 - self.beta2 ** t)

        self.weights -= self.learning_rate * m_t_hat / (np.sqrt(v_t_hat) + self.epsilon)
        self.bias -= self.learning_rate * m_t_bias_hat / (np.sqrt(v_t_bias_hat) + self.epsilon)

    def predict(self, X):
        return self._predict(X) #computes the predicted target values using the trained model.
```

#observations:

1. Standard SGD: Standard SGD updates the parameters using the gradient of the loss with respect to the parameters. It can take longer to converge because it does not incorporate any momentum or adaptive learning rate mechanisms to accelerate convergence or handle varying gradients efficiently. Without these features, it might require more iterations to find the minimum loss.

2. SGD with Adam: Adam combines the advantages of adaptive learning rates. It adapts the learning rates for each parameter based on past gradients and squared gradients, which can lead to faster convergence.

3. SGD with Momentum: SGD with momentum typically converges faster than both standard SGD and SGD with Adam because the momentum term helps accelerate convergence by accumulating gradients from previous iterations, which helps the optimizer navigate through flat regions and escape local minima more efficiently. This acceleration can lead to the shortest convergence time among the three algorithms.
